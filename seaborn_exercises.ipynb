```python
# 4. Using the chipotle dataset from the previous exercise, create 
# a bar chart that shows the 4 most popular items and the revenue 
# produced by each.
```

```python
import env
```

```python
url = env.get_db_url('chipotle')
```

```python
chipotle = pd.read_sql('select * from orders;', url)
```

```python
chipotle
```

```python
chipotle['revenue'] = ''
```

```python
chipotle.item_price = chipotle.item_price.str.strip('$').str.strip(',')
```

```python
chipotle.dtypes
```

```python
chipotle.item_price = chipotle.item_price.astype(float)
```

```python
chipotle.item_name.value_counts().head(4) 
```

```python
most_pop_items = chipotle.groupby(['item_name']).agg({'item_price': sum}).sort_values(by = 'item_price', ascending=False).head(4)
```

```python
most_pop_items = most_pop_items.reset_index()
```

```python
most_pop_items
```

```python
sns.barplot(data=most_pop_items, x='item_name', y='item_price')
plt.title('Four Most Popular Items')
plt.xlabel('Item Name')
plt.ylabel('Total Revenue')
plt.show()
```

```python
# 5. Load the sleepstudy data and read it's documentation.
```

```python
sleepstudy = data('sleepstudy')
```

```python
sleepstudy
```

```python
data('sleepstudy', show_doc = True)
```

```python
# Use seaborn to create a line chart of all the individual 
# subject's reaction times and a more prominant line showing the 
# average change in reaction time.
```

```python
sns.lineplot(data=sleepstudy, x='Days', y='Reaction', hue='Subject')
plt.title('Changes in Reaction Time with Sleep Deprivation')
plt.xlabel('Days with Only 3 Hours of Sleep')
plt.ylabel('Time to React in ms')
plt.show()
```

```python
# Lets force Seaborn to interpret subjects less numerically 
```

```python
sleepstudy['subject_cat'] = 'Subject: ' + sleepstudy.Subject.astype(str)
```

```python
sns.relplot(data=sleepstudy, x='Days', y='Reaction', hue='subject_cat', kind='line')
plt.title('Changes in Reaction Time with Sleep Deprivation')
plt.xlabel('Days with Only 3 Hours of Sleep')
plt.ylabel('Time to React in ms')
plt.show()
```

```python
# now lets get the mean line
```

```python
plt.figure(figsize=(16, 10))

sns.relplot(data=sleepstudy, 
            x='Days', 
            y='Reaction', 
            hue='subject_cat', 
            kind='line')

sns.lineplot(data=sleepstudy, 
             x='Days', 
             y='Reaction', 
             legend=False, 
             color='black',
             estimator='mean')

plt.title('Changes in Reaction Time with Sleep Deprivation')
plt.xlabel('Days with Only 3 Hours of Sleep')
plt.ylabel('Time to React in ms')
plt.annotate('Average of all subjects', (1, 250))

plt.show()
```
