```python
# 2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

# y=âˆšx
# y=x^3
# y=2^x
# y=1/(x+1)

# You can use functions from the math module to help implement some of the equations above.
```

```python
plt.figure(figsize = (10,8))

x = range(0, 21)
x = list(x)


plt.subplot(221)
y1 = [ (math.sqrt(n)) for n in x ]
plt.plot(x, y1)
plt.title('The Squareroot of X')

plt.subplot(222)
y2 = [ (n ** 3) for n in x ]
plt.plot(x, y2)
plt.title('X Cubed')

plt.subplot(223)
y3 = [ (2 ** n) for n in x ]
plt.plot(x, y3)
plt.title('2 to the Power of X')

plt.subplot(224)
y4 = [ 1 / (n + 1) for n in x ]
plt.plot(x, y4)
plt.title('X Plus One Divided by 1')

```

```python
plt.figure(figsize = (8, 6))

x = [ n * 1 for n in range(0, 21) ]

y1 = [ (math.sqrt(n)) for n in x ]

y2 = [ (n ** 3) for n in x ]

y3 = [ (2 ** n) for n in x ]

y4 = [ 1 / (n + 1) for n in x ]

plt.plot(x, y1, c='#30a2da', label='$\sqrt{x}$', alpha=1.0)
plt.plot(x, y2, c='orange', label='$x^{3}$', alpha=.8)
plt.plot(x, y3, c='green', label='$2^{x}$', alpha=.8)
plt.plot(x, y4, c='purple', label='1/(x+1)$', alpha=.8)

plt.title('Arithmetic of X')

plt.legend()

plt.ylim(0, 100)
```
