```python
# 2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

# y=âˆšx
# y=x^3
# y=2^x
# y=1/(x+1)

# You can use functions from the math module to help implement some of the equations above.
```

```python
x = range(0, 21)
x = list(x)


plt.subplot(221)
y1 = [ (math.sqrt(n)) for n in x ]
plt.plot(x, y1)
plt.title('The Squareroot of X')

plt.subplot(222)
y2 = [ (n ** 3) for n in x ]
plt.plot(x, y2)
plt.title('X Cubed')

plt.subplot(223)
y3 = [ (2 ** n) for n in x ]
plt.plot(x, y3)
plt.title('2 to the Power of X')

plt.subplot(224)
y4 = [ 1 / (n + 1) for n in x ]
plt.plot(x, y4)
plt.title('X Plus One Divided by 1')

```
