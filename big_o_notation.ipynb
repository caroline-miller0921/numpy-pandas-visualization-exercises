```python
from matplotlib import pyplot as plt
import pandas as pd
from pydataset import data

## Data generation
import numpy as np
import random

## Mathematical functions
import math

## Random seed value to reproduce results
np.random.seed(123)


# Set retina for higher resolution display
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
# 5. Make a new Jupyter notebook named big_o_notation.ipynb

 # Title your chart "Big O Notation"
# Label your x axis "Elements"
# Label your y axis "Operations"
# Label your curves or make a legend for the curves
# Use LaTex notation where possible

# Curves to graph: 

plt.figure(figsize = (8, 6))

x = range(1, 21)
x = list(x)

# y = 0n + 1
y1 = [ (0 * n) + 1 for n in x]
plt.plot(x, y1, c='red', label='$y = 0_n')


plt.title('Big O Notation: y = 0n + 1')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()

plt.ylim(0, 5)

```

```python

plt.figure(figsize = (8, 6))

x = range(1, 21)
x = list(x)

# y = log(n)
y2 = [ math.log(n) for n in x ]
plt.plot(x, y2, c='green', label='$y = log(n)$')

plt.title('Big O Notation: y = log(n)')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()

plt.ylim(0, 4)
```

```python
plt.figure(figsize = (8, 6))

x = range(1, 21)
x = list(x)

# y = n
y3 = [ n for n in x ]
plt.plot(x, y3, c='magenta', label='$O(n)$')

plt.title('Big O Notation: y = n')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()

plt.ylim(0, 25)
```

```python
plt.figure(figsize = (8, 6))

x = range(1, 21)
x = list(x)

# y = n * log(n)
y4 = [  n * math.log(n) for n in x  ]
plt.plot(x, y4, c='black', label='$y = O(n log n)$')

plt.title('Big O Notation: y = n * log(n)')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()

plt.ylim(0, 60)
```

```python
plt.figure(figsize = (8, 6))

x = range(1, 21)
x = list(x)

# y = n^2
y5 = [  n ** 2 for n in x  ]
plt.plot(x, y5, c='lavender', label='$y = n^2$')

plt.title('Big O Notation: y = n^2')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()

plt.ylim(0, 400)
```

```python
plt.figure(figsize = (8, 6))

x = range(1, 21)
x = list(x)

# y = 2^n
y6 = [  2 ** n for n in x  ]
plt.plot(x, y6, c='blue', label='$y = 2^n$')

plt.title('Big O Notation: y = 2^n')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()

plt.ylim(0, 1100000)
```

```python
plt.figure(figsize = (8, 6))

x = range(1, 21)
x = list(x)

# y = n!
y7 = [ math.factorial(n) for n in x  ]
plt.plot(x, y7, c='cyan', label='$y = n!$')

plt.title('Big O Notation: y = n!')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()
```
